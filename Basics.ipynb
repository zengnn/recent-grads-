```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
#number of rows
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
#dropping rows containing missing values
recent_grads = recent_grads.dropna(axis = 0)
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
recent_grads.plot(x='Men', y='Median', kind='scatter')
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
#total number of people in a major vs income
recent_grads.plot(x = 'Total', y = 'Median', kind = 'scatter')
recent_grads.plot(x = 'Total', y = 'P25th', kind = 'scatter')
recent_grads.plot(x = 'Total', y = 'P75th', kind = 'scatter')
```

The people who graduated with a more popular major seem to have an average to lower than average income
This is consistant across the median, 25th quantile, and 75th quantile income. 

```python
#percentage of women vs. median income
recent_grads.plot(x = 'ShareWomen', y = 'Median', kind = 'scatter')
recent_grads.plot(x = 'ShareWomen', y = 'P25th', kind = 'scatter')
recent_grads.plot(x = 'ShareWomen', y = 'P75th', kind = 'scatter')
```

The students that majored in subjects with majority female make less money than if they majored in a subject with a less share of women. This is consistant across the median, 25th quantile, and 75th quantile income. 

```python
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter')
```

There's no strong correlation between the number of full time employees and median salary.
